```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Graph 6 is based on Graph 3 (bfs_candidate_2) but uses static variables which should in theory reduce initialisation and allocation overhead.
However, in reality, there is little to no difference in performance. So Graph 3 will be used due to its safety.
