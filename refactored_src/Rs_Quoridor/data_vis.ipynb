```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df1 = pd.read_csv("data/2023-09-12_11-16-22-10.csv") # BFS
df2 = pd.read_csv("data/2023-09-12_11-16-20-10.csv") # BFS C1 with Vec and Arr
df3 = pd.read_csv("data/2023-09-12_11-16-18-10.csv") # BFS C2 with Arr and Arr 
df4 = pd.read_csv("data/2023-09-12_11-16-16-10.csv") # BFS C3 with Vec and Vec
df5 = pd.read_csv("data/2023-09-12_11-16-14-10.csv") # BFS C4 with Arr and Vec
df6 = pd.read_csv("data/2023-09-12_11-44-53-10.csv") # BFS C5 with static Arr and static Arr

```

```python
plt.figure(dpi=250)
df1.query('type_board == 1')['time_taken'].head(100).plot(kind='line', title='Time taken vs Actions taken', legend=True)
df1.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
df2.query('type_board == 1')['time_taken'].head(100).plot(kind='line', legend=True)
df2.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
df3.query('type_board == 1')['time_taken'].head(100).plot(kind='line', legend=True)
df3.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
df4.query('type_board == 1')['time_taken'].head(100).plot(kind='line', legend=True)
df4.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
df5.query('type_board == 1')['time_taken'].head(100).plot(kind='line', legend=True)
df5.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
plt.legend(['StaticGraph1', 'DynamicGraph1', 'StaticGraph2', 'DynamicGraph2', 'StaticGraph3', 'DynamicGraph3','StaticGraph4', 'DynamicGraph4', 'StaticGraph5', 'DynamicGraph5'])
```

In pathfinding, there is a frontier which is either a stack, queue or "priority queue" based on which algorithm is in use.

* *Graph1 - BFS using VecDeque
* *Graph2 - BFS using Vec instead of VecDeque
* *Graph3 - BFS using Arr instead of VecDeque
* *Graph4 - BFS using Vec instead of VecDeque with vec for path gen
* *Graph5 - BFS using Arr instead of VecDeque with vec for path gen

Graph1 and Graph3 have the best performance - as a result, Graph2, Graph4, Graph5 will be ignored below. Graph3 has minutely more performance according to the graph.





```python
plt.figure(dpi=250)
df1.query('type_board == 1')['time_taken'].head(100).plot(kind='line', title='Time taken vs Actions taken', legend=True)
df1.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
df2.query('type_board == 1')['time_taken'].head(100).plot(kind='line', legend=True)
df3.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
df6.query('type_board == 1')['time_taken'].head(100).plot(kind='line', legend=True)
df6.query('type_board == 2')['time_taken'].head(100).plot(kind='line', legend=True)
plt.legend(['StaticGraph1', 'DynamicGraph1','StaticGraph3', 'DynamicGraph3','StaticGraph6', 'DynamicGraph6'])
```

Graph 6 is based on Graph 3 (bfs_candidate_2) but uses static variables which should in theory reduce initialisation and allocation overhead.
However, in reality, there is little to no difference in performance.
