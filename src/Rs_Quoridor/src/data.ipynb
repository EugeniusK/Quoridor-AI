```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/2023-09-22_12-01-16-500.csv")
plt.figure(dpi=120)
df.query('mode== 1').head(10000).groupby('num_walls')['time_taken'].mean().plot(kind='line', title='Time taken vs Actions taken', legend=True)
df.query('mode== 2').head(10000).groupby('num_walls')['time_taken'].mean().plot(kind='line', legend=True)
df.query('mode== 3').head(10000).groupby('num_walls')['time_taken'].mean().plot(kind='line', legend=True)
df.query('mode== 4').head(10000).groupby('num_walls')['time_taken'].mean().plot(kind='line', legend=True)

plt.legend(['BFS', 'DFS', 'GBFS', 'A*'])
```

```python
df = pd.read_csv("../data/2023-09-22_12-01-16-500.csv")
plt.figure(dpi=120)
df.query('mode== 1  and num_actions <= 50').groupby('num_actions')['time_taken'].mean().plot(kind='line', title='Time taken vs Actions taken', legend=True)
df.query('mode== 2 and num_actions <= 50').groupby('num_actions')['time_taken'].mean().plot(kind='line', legend=True)
df.query('mode== 3 and num_actions <= 50').groupby('num_actions')['time_taken'].mean().plot(kind='line', legend=True)
df.query('mode== 4 and num_actions <= 50').groupby('num_actions')['time_taken'].mean().plot(kind='line', legend=True)

plt.legend(['BFS', 'DFS', 'GBFS', 'A*'])
df.describe()
```
